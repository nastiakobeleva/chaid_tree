Загрузка и подготовка данных (этап 1)

```python
import pandas, randan, re
```

```python
%%time
#достать файлы
path = r'C:\Users\vgv\Downloads\Мага'
path = re.sub(r'\\',r'\\',path)
data = pandas.read_csv(f'{path}\\Матрица_Релевантность.csv', index_col=0)

dpndnt = 'Релевантность_теме_исследования' #зависимая переменная
```

```python
#удалить строки датафрейма с пустыми  документами
data = data[(data['Предообработанный_текст'].notna()) & (data['Предообработанный_текст'] != '')]
data
```

```python
#временно поделить датафрейм на два, чтобы заменить все частоты больше 0 на единицу
data_docS = data[['Релевантность_теме_исследования', 'snippet.textOriginal', 'Предообработанный_текст']]
data_tokenS = data.drop(['Релевантность_теме_исследования', 'snippet.textOriginal', 'Предообработанный_текст'], axis=1)
tokenS = data_tokenS.columns
```

```python
#замена частот
data_tokenS[data_tokenS >= 1] = 1
```

```python
#проверка
print(f'{(data_tokenS == 0).sum().sum()} ячеек с нулевыми частотами')
print(f'{(data_tokenS == 1).sum().sum()} ячеек с единичными частотами')
print(f'{(data_tokenS > 1).sum().sum()} ячеек с частотами выше единичных')
```

```python
#снова соединить части датафрейма
data = pandas.concat([data_docS, data_tokenS], axis=1)
data
```

```python
#описательная статистика
print(f"{(data_docS[dpndnt].notna()).sum()} валидных значений зависимой переменной")

display('Распределение зависимой переменной на всех размеченных данных (в долях)',
       data_docS[dpndnt].value_counts(normalize=True))
```

```python
data_tagged = data[data[dpndnt].notna()] #учительский датафрейм
data_untagged = data[data[dpndnt].isna()] #неразмеченный датафрейм
display(data_tagged, data_untagged)
```

```python
#разделить учительский датафрейм по строкам на обучающую (75%) и тестовую (25%) выборки
data_tagged_training = data_tagged.sample(frac=0.75, random_state=1) #обучающая выборка

data_tagged_test = data_tagged.drop(data_tagged_training.index) #тестовая выборка
display('Обучающая выброрка:', data_tagged_training, 'Тестовая выборка:', data_tagged_test)
```

```python
display('Распределение зависимой переменной в обучающей выборке (в долях)',
       data_tagged_training[dpndnt].value_counts(normalize=True))
```

```python
data_tagged_training[tokenS]
```

```python
data_tagged_training[tokenS].var() #вывести дисперсию
```

```python
tokenS_out = list(data_tagged_training[tokenS].var()[data_tagged_training[tokenS].var() == 0].index)
print(f'Токены без дисперсии -- на исключение: {tokenS_out}')
```

```python
print(f'Проверка: дисперсия токена "{tokenS_out[0]}" равна', data_tagged_training[tokenS].var()[tokenS_out[0]])
```

```python
tokenS_in = list(data_tagged_training[tokenS].var()[data_tagged_training[tokenS].var() != 0].index)
print(f'Остаются {len(tokenS_in)} токенов: {tokenS_in}')
```

2 этап: обучение Chaid на размеченных вручную документах

*обучить дерево

```python
%%time
chaid = randan.tree.CHAIDClassifier(max_depth=None,
                                   min_child_node=10,
                                   min_parent_node=0).fit(data_tagged_training,
                                                          dpndnt,
                                                          tokenS_in,
                                                          test_data = data_tagged_test)
```

```python
predictionS = chaid.predict(data_untagged)
print(f'{predictionS.sum()} релевантных документов среди размеченных деревом')
```

```python
data_untagged = pandas.concat([predictionS, data_untagged], axis=1)
data_untagged
```
